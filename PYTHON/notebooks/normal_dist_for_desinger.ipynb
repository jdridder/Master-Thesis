```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import scienceplots
import matplotlib.cm as cm

def set_mpt_settings(font_size: float = 16, figsize=(10, 6)):
    plt.style.use("science")
    matplotlib.rcParams.update(
        {
            "lines.linewidth": 2,
            "lines.markeredgecolor": "black",
            "lines.markeredgewidth": 0.5,
            "font.size": font_size,
            "axes.labelsize": "large",
            "figure.figsize": figsize,
            "axes.grid": False,
            "lines.markersize": 10,
            "axes.unicode_minus": False,
            "ps.fonttype": 42,  # Avoid type 3 fonts
            "pdf.fonttype": 42,  # Avoid type 3 fonts
            "pgf.texsystem": "xelatex",  # oder xelatex/lualatex
            "text.usetex": True,
            "text.latex.preamble": r"\usepackage{bm} \usepackage{amsmath} \usepackage{amsfonts}",
            "font.family": "serif",
            "font.serif": ["Computer Modern Roman"],
        }
    )
set_mpt_settings()

def make_colors(n_colors: int, alpha: float = 1, cmap: str = "viridis"):
    cmap = cm.get_cmap(cmap)
    return cmap(np.linspace(0, 1, n_colors), alpha=alpha)
```

```python
fig, ax = plt.subplots(1, figsize=(1, 13.68))

s = 1.7
fun = lambda x: 1 / ((2*np.pi)**0.5 * s) * np.exp(-0.5 * x**2/s**2)

x_arr = np.linspace(-5, 5, 100)
pdf = fun(x_arr)

ax.fill_between(pdf, x_arr, linewidth=5, color=make_colors(4)[1])
ax.set_yticklabels(labels=[])
ax.set_xticklabels(labels=[])

plt.savefig("normal_dist.pdf")
```
